```python
# setting genius api features
genius.verbose = False  # turn off status messages
genius.remove_section_headers = True  # remove section headers (ex. chorus)
genius.timeout = 15  # timeout before quitting (increase when wifi is slower)
genius.retries = 2 # retry 2 times before skipping

# note: lyricsgenius AUTOMATICALLY handles staggering requests so no need for time delay
```

```python
# define filepath
folder = 'datasci112_lyrics'
os.makedirs(folder, exist_ok=True)
csv_filename = 'genius_data.csv'
```

```python
# function to get lyrics and add them to df
def get_lyrics(songs, csv_name):
  filepath = os.path.join(folder, csv_name)

  with open(filepath, 'w', encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(['artist', 'song', 'lyrics', 'album', 'release_year', 'url'])

# trying each artist, title tuple, adding found data and lyrics
    for artist, title in songs:
      try:
        song = genius.search_song(title, artist)
        album = song.album if song.album else "unknown album"
        release_year = song.year if song.year else "unknown year"
        song_url = song.url if song.url else "n/a"
        writer.writerow([artist, title, song.lyrics, album, release_year, song_url])
        print(f"{title} by {artist} saved")
      except:
        print(f"{title} by {artist} not found")
        continue
```

```python
# creating a list of artist, song tuples
song_list = df['tuple_col'].tolist()

song_list
```

```python
# getting lyrics for all artist, song pairs
get_lyrics(song_list, csv_filename)
```
