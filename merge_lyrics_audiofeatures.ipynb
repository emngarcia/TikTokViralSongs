```python
import pandas as pd
```

```python
# reading in df with genius lyrics, track names, and artist names
df_lyrics = pd.read_csv('genius_data.csv')

# preprocessing before merging
df_lyrics = df_lyrics.rename(columns={'song': 'track_name', 'artist': 'artist_name'})

df_lyrics_nodupes = df_lyrics.drop_duplicates(subset=['track_name', 'artist_name'])
df_lyrics_nodupes = df_lyrics_nodupes.reset_index(drop=True)

df_lyrics_nodupes
```

```python
# reading in tiktok song audio features
df_song_info = pd.read_csv('tiktok_data.csv')

# preprocessing before merging
df_song_info = df_song_info.drop(columns=['Unnamed: 0'])

df_song_info
```

```python
# merging left to include duplicates
df_merged_left = df_song_info.merge(df_lyrics_nodupes, on = ['track_name', 'artist_name'], how = 'left')

df_merged_left
```

```python
# merging right to not include duplicates, by default merges to first observation of each key, so saves for earliest year the song was popular
df_merged_right = df_song_info.merge(df_lyrics_nodupes, on = ['track_name', 'artist_name'], how = 'right')

df_merged_right
```
